#### Import Libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn import metrics
import seaborn as sns
```

##### plot sales

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=data['TV'],y=data['Sales'])
plt.show()
plt.figure(figsize=(8,5))
sns.scatterplot(x=data['Radio'],y=data['Sales'])
plt.show()
plt.figure(figsize=(8,5))
sns.scatterplot(x=data['Newspaper'],y=data['Sales'])
plt.show()
```

##### split data

```python
train,test = train_test_split(data,test_size=0.30)
```

```python
train_x=train[['TV', 'Radio', 'Newspaper']]
train_y=train.Sales

test_x=train[['TV', 'Radio', 'Newspaper']]
test_y=train.Sales
```

```python
train_x.head
```

```python
train_y.head
```

##### Train model

```python
model= linear_model.LinearRegression()
model.fit(train_x,train_y)
predict=model.predict(test_x)
metrics.r2_score(predict,test_y)
```

##### Regression Graph

```python
plt.figure(figsize=(10,5))
sns.regplot(x=test_y,y=predict,scatter_kws={'color':'Blue'})
```
